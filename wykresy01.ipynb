```python
x = np.linspace(0, 2*np.pi, 42)

f1 = 5 * np.sin(x)
f2 = 2 * np.sin(2*x)
f3 = 1 * np.sin(3*x)

plt.plot(x, f1)
plt.plot(x, f2, 'ro')
plt.plot(x, f3, 'g--')
```

```python
x = np.linspace(0, 2*np.pi, 42)

f1 = x**2
f2 = f1 + 5
f3 = 2 * f1

plt.plot(x, f1, 'r-', label='f1')
plt.plot(x, f2, 'b-', label='f2')
plt.plot(x, f3, 'g-', label='f2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.show()
```
